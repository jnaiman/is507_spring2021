## 1. Basic $\chi^2$ things

Let's start with our usual - plotting these distributions for a few values of the DOF parameter:

```r
x = seq(0, 50, length=200)
dist = dchisq(x, df=3)

plot(x, dist, type='l', ylab='Probability Distribution of Chi-Sq', xlab='Chi-Squared Value')
```

Ok, now lets loop through and plot a few different degrees of freedom:

```r
dfs = c(5, 10, 30)
```

Let's make them all different colors too:

```r
colors = c("red", "blue", "magenta")
```

Now let's replot:

```r
plot(x, dist, type='l', ylab='Probability Distribution of Chi-Sq', xlab='Chi-Squared Value')
for (i in 1:length(dfs)){
  lines(x, dchisq(x,df=dfs[i]), col=colors[i])
}
```

Let's add another little legend for making things look nice making sure we include the first one:

```r
dfs = c(3, dfs) # append to dfs list
colors = c("black",colors) # append to colors list
```

```r
plot(x, dist, type='l', ylab='Probability Distribution of Chi-Sq', xlab='Chi-Squared Value')
for (i in 1:length(dfs)){
  lines(x, dchisq(x,df=dfs[i]), col=colors[i])
}

# legend
legend("topright", as.character(dfs), col=colors, lwd=2)
```

We can see that the distribution slides to higher chi-squared values for higher degrees of freedom.

It even looks like as DOF increases it approaches a normal curve...

Let's plot something to guide our eye...

```r
plot(x, dist, type='l', ylab='Probability Distribution of Chi-Sq', xlab='Chi-Squared Value')
for (i in 1:length(dfs)){
  lines(x, dchisq(x,df=dfs[i]), col=colors[i])
}

lines(x, dnorm(x,mean=29,sd=7),col="green")

# legend
legend("topright", as.character(dfs), col=colors, lwd=2)
```

This green normal curve doesn't overly the highest DOF curve, but it looks close.

This is actually a feature of chi-squards - as the DOF increases, it approaches a normal curve. (This once again comes from the Central Limit Theorem). Again, this is why normals are so famous!

Let's go back to just one plot:

Say we have some made up distribution that has 8 DOF, and when we calculate our chi^2, we find a value of 10. What is the p-value of this test statistic?

```r
df = 8
test_stat = 10
```

Let's start by plotting:

```r
x = seq(0, 50, length=200)
plot(x, dchisq(x, df=df), type='l', 
     ylab='Probability Distribution of Chi-Sq', 
     xlab='Chi-Squared Value')
```

Turns out, just like for the normal & binomial distribution we can calculate percentages with a "p" function:

```r
pvalue = 1-pchisq(test_stat, df=df) 
print(pvalue)
```

Also can do:

```r
pvalue = pchisq(test_stat, df=df, lower.tail=FALSE)
print(pvalue)
```

Thinking: why the same?

Now lets plot what this corresponds to on our graph:

```r
x2=seq(test_stat,50,length=200)
y2=dchisq(x2,df=df)
```

```r
plot(x, dchisq(x, df=df), type='l', 
     ylab='Probability Distribution of Chi-Sq', 
     xlab='Chi-Squared Value')

draw_polygon(x2,y2)
```

## 2. Labby's dice

First let's make a vector holding each die roll:

```r
die_roll = c(1, 2, 3, 4, 5, 6)
```

Now lets plug in what Labby found for the observed rolls:

```r
obs = c( 53222, 52118, 52465, 52338, 52244, 53285 )
```

From probability theory:

```r
# recall
number_of_dice_rolled_at_once = 12 # the number of die on the table
number_of_rolls = 26306 # How many rolls the machine did
```

Let's calculate the expected value in each bin:

```r
expected_value_of_each_side = number_of_dice_rolled_at_once * number_of_rolls/6
```

Now, since this is true for all sides of the die we can for-loop over this:

```r
expected = c()
for (i in 1:6){
  expected = c(expected,expected_value_of_each_side)
}
```

Before we get to calculations, lets plot these results:

```r
barplot(obs, names.arg=die_roll)
```

Let's overplot what we expect from "theory":

```r
barplot(obs, names.arg=die_roll)
abline(h=expected_value_of_each_side, col="blue")
```

Let's zoom our plot a bit to check it out:

```r
barplot(obs, names.arg=die_roll, ylim=c(50000,55000))
abline(h=expected_value_of_each_side, col="blue")
```

We can see at the very top there, indeed it looks like 1 & 6 are more frequent then the others.

Now, let's calculate each element of the total chi squared test statistic:
  * recall $\chi^2$ = sum$\left ( \frac{(observation-expected)^2}{expected} \right)$

```r
chi_vals = (obs - expected)**2/expected
chi_vals
```

```r
test_stat_chi = sum(chi_vals)
print(test_stat_chi)
```

Now, we need to compare this to the appropriate chi-squared distribution:

```r
df = length(obs)-1
pvalue = pchisq(test_stat_chi,df=df, lower.tail=FALSE)
print(pvalue)
```

We see that this pvalue is much less than our typical level of confidence of $\alpha$ = 0.05 so we can reject the null hypothesis, i.e. the observed die rolls are indeed different than the expected values and so, we say this tests provides convincing evidence that the dies are biased.

Now we have "levels" associated with things:

Now, let's translate the different "levels" from 1, 2, NA to actual values -- look at help(election) to copy paste:

```r
levels(election$VOTEF) <- c("Gore","Bush","Other")
levels(election$EDUC) <- c("<8th", "9-11 grades","HS Diploma", ">12 years schooling", 
                           "Junior College Degree", "BA/BS", "Adv. deg.")
levels(election$GENDER) <- c("Male", "Female")
levels(election$PARTY) <- c("Strong Democrat", "Weak Democrat", 
                            "Independent-Democrat", "Independent-Independent",
                            "Independent-Republican", "Weak Republican", "Strong Republican")
```

```r
head(election$GENDER)
```

```r
head(election)
```

There are a lot of NAs!  Let's take them out:

```r
election2 = na.omit(election[,c("VOTEF","PARTY","EDUC","GENDER")])
```

Let's ask something that might be very intuative - is there a relationship between voting party and who they voted for?  Let's state the null & alternative hypothesis:

$H_0$: There is no relationship between the party affiliation (7 levels) and voting results (Bush, Gore, Other) in the population.

$H_A$: There is a relationship between the party affiliation (7 levels) and voting results (Bush, Gore, Other) in the population.


First, let's verify conditions for using chi-squared

**Independence:** probably, but we aren't sure, there are also probably biases in how this population was sampled.

What does our data look like?

```r
electable = table(election2$PARTY, election2$VOTEF)
```

```r
electable
#print(electable)
```

If our null hypothesis is true than we expect an even distribution of counts in the three categories of "Gore", "Bush", "other".

By eye we can see that there is *not* an even distribution.  

Note: this is going to involve a chi-squared test for a 2D table.  Our table degrees-of-freedom:
  * table df: (# rows - 1) * (#columns - 1) = (7 - 1) * (3 - 1) = 12

**Enough samples?**: Let's see if the $\chi^2$ is going to work:

```r
#print(chisq.test(electable)$expected)
chisq.test(electable)$expected
```

We get an error - this is because in "Other" & "I-I" we have < 5 samples as our expected value.

In reality, $\chi^2$ won't be as accurate here & we should do another test, but let's proceed anyway.

```r
#print(chisq.test(electable))
chisq.test(electable)
```

Note our p-value is SUPER small so we can reject the null hypothesis and assume that party affilation & who they voted for are related which makes sense with our initial intuition.

### Extra example 

Let's do another quick one: Is gender & who they voted for related.

$H_0$: There is no relationship between gender (2 levels) and voting results (Bush, Gore, Other) in the population.

$H_A$: There is a relationship between gender (2 levels) and voting results (Bush, Gore, Other) in the population.

Let's check out table to look at this:


```r
gentable = table(election2$VOTEF, election2$GENDER)
```

```r
#print(gentable)
gentable
```

In this case, its sort of hard to see - *maybe* more women voted for Gore?

But it's hard to tell, let's use a $\chi^2$ test to find out!


```r
#print(chisq.test(gentable))
chisq.test(gentable)
```

Here we see that our pvalue is 0.0015 < 0.05 so we can reject the null hypothesis that these two variables are not related, and gender and voting have a relationship.
